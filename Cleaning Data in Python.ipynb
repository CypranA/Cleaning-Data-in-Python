# Data Range constraints

How to deal without range:
. Dropping data: only drop data when a small proportion of your dataset is affected by out of range values
. Setting custom minimums and maximums
. Treat as missing and impute

```python
movies = pd.read_csv('movies.csv')
movies

#rating is any integer between 1 - 5
#some movies have an average rating exceeding 6 which needs to be fixed
```

# How to find Duplicate Rows

Using only the .duplicate() method will not always provide an accurate result of rows with duplicate. Having duplicates in some columns does not mean that the rows are duplicate.

A more accurate way to find duplicate rows are as follow:
1. The .duplicated() method
   Subset: List of column names to check for duplication
   Keep: Whether to keep first ('first'), last ('last'), or all (False) duplicate values
   

```python
#column names to check for duplication

column_names = ['col_name1', 'col_name2', 'col_name3']
duplicates = df.duplicated(subset = column_names, keep = False)
df[duplicates]

#sorting the duplicates to get a better eyeview
#sorting duplicates by first column name ('col_name1')

df[duplicates].sort_values(by = 'col_name1') #
```

# Comparing Strings: Record Linkage 

Record linkage is the act of linking data from different sources regarding the same entity. Merging data has to be done carefully to avoid duplication. 

The process of record linkage is as follows:

1. Extract 2 or more data. eg, Data A and Data B
2. Generate pairs
3. Compare pairs
4. Score pairs
5. Link Data

```python
census_A = pd.read_csv('census_A.csv', index_col=0)
census_A
```

```python
census_B = pd.read_csv('census_B.csv', index_col=0)
census_B
```

```python
#generating pairs

import recordlinkage

#create indexing object
indexer = recordlinkage.Index()

#generate pairs blocked on state
indexer.block('state')
pairs = indexer.index(census_A, census_B)
pairs
```

```python
#comparing the dataframes
#create a compare object

compare_cl = recordlinkage.Compare()

#find exact matches for pairs of date_of_birth and state
compare_cl.exact('date_of_birth', 'date_of_birth', label='date_of_birth')
compare_cl.exact('state', 'state', label='state')

#find similar matches for pairs of surname and address_1 using string similarity
compare_cl.string('surname', 'surname', threshold=0.85, label='surname')
compare_cl.string('address_1', 'address_1', threshold=0.85, label='address_1')

#find matches
potential_matches = compare_cl.compute(pairs, census_A, census_B)
potential_matches
```

```python
#isolate matches with matching values for 2 or more columns
matches = potential_matches[potential_matches.sum(axis=1) >= 2]
matches
```

```python
#linking dataframes

#get indices
matches.index

#get indices from census_B only
duplicate_rows = matches.index.get_level_values(1)
print(census_B.index)
```

```python
#find duplicates in census_B
census_B_duplicates = census_B[census_B.index.isin(duplicate_rows)]

#finding new rows in census_B
census_B_new = census_B[~census_B.index.isin(duplicate_rows)]

#link the dataframes
full_census = census_A.append(census_B_new)

#generate potential matches
potential_matches1 = compare_cl.compute(full_census, census_A, census_B)

#isolate matches with matching values for 3 or more columns
matches = potential_matches[potential_matches.sum(axis=1) >= 2]

#get index for matching census_B only
duplicate_rows = matches.index.get_level_values(1)

#finding new rows in census_B
census_B_new = census_B[~census_B.index.isin(duplicate_rows)]

#link/append the dataframes
full_census = census_A.append(census_B_new)
```
